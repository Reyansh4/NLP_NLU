# **IMPORTING THE REQUIRED LIBRARIES**

```python
from summarizer import Summarizer
from datasets import load_dataset
from rouge import Rouge
import warnings
```

# **IMPORTING THE DATASET**

Here we are using the only the test dataset because we are already using a pretrained model. So we dont need the training and valid dataset so we can skip training the model

```python
test_dataset = load_dataset('cnn_dailymail','3.0.0', split = 'test')
```

```python
print("The length if the test dataset is ",len(test_dataset))
```

# **LOAD THE MODEL**

```python
bert_model = Summarizer()
```

```python
Summaries = []

for body in test_dataset['article'][:10]:
    bert_summary = ''.join(bert_model(body, min_length=20))
    Summaries.append(bert_summary)
```

# **EVALUATION BY ROGUE SCORE**

```python
rouge = Rouge()
scores = rouge.get_scores(Summaries, test_dataset["highlights"][:10], avg=True)

print(f"ROUGE-1: {scores['rouge-1']['f']:.4f}")
print(f"ROUGE-2: {scores['rouge-2']['f']:.4f}")
print(f"ROUGE-L: {scores['rouge-l']['f']:.4f}")
```
