```python
from tqdm import tqdm
from datasets import load_dataset
from transformers import AutoTokenizer,DataCollatorWithPadding

tqdm.pandas()

dataset = load_dataset("AmazonScience/massive")
```

```python
checkpoint = "cartesinus/xlm-r-base-amazon-massive-slot"   #cartesinus/xlm-r-base-amazon-massive-domain
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

def tokenize_function(example):
    return tokenizer(example["utt"], truncation=True)
```

```python
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
import json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import transformers
import torch
from transformers import AutoTokenizer,AutoModelForSequenceClassification
```

```python
json_file_path = '/content/drive/MyDrive/Colab Notebooks/IndoML IITB contest/surprise.data'

train_data = []

with open(json_file_path, 'r') as json_file:
    for line in json_file:
      data = json.loads(line)
      train_data.append(data)
```

```python
json_file_path = '/content/drive/MyDrive/Colab Notebooks/IndoML IITB contest/surprise.solution'

solutions = []
with open(json_file_path, 'r') as json_file:
    for line in json_file:
      solution = json.loads(line)
      solutions.append(solution)
```

```python
print(solutions[0])
```

```python
intent_map = {item['indoml_id']: item['intent'] for item in solutions}

indoml_ids = [item['indoml_id'] for item in train_data]
intents = [intent_map[indoml_id] for indoml_id in indoml_ids]

utt = [item['utt'] for item in train_data]

num_classes = len(set(intents))
print("The no of classes are : ", num_classes)

train_data,test_data,train_labels,test_labels = train_test_split(utt, intents, test_size = 0.33, random_state = 42, stratify = intents)
```

```python
model_name = "cartesinus/xlm-r-base-amazon-massive-domain"
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_classes, ignore_mismatched_sizes = True)
```

```python
train_encodings = tokenizer(
    train_data,
    truncation=True,
    padding=True,
    max_length=32,
    return_tensors='pt'
)

test_encodings = tokenizer(
    test_data,
    truncation=True,
    padding=True,
    max_length=32,
    return_tensors='pt'
)
```

```python
label_map = {intent: i for i, intent in enumerate(set(train_labels))}
train_labels = [label_map[label] for label in train_labels]
test_labels = [label_map[label] for label in test_labels]
```

```python
print(label_map)
```

```python
class IntentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = IntentDataset(train_encodings, train_labels)
test_dataset = IntentDataset(test_encodings, test_labels)
```

```python
train_dataset[0]
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',
    per_device_train_batch_size=15,
    per_device_eval_batch_size=15,
    evaluation_strategy="epoch",
    logging_steps=41,
    eval_steps=41,
    save_total_limit=5,
    learning_rate=475e-7,
    num_train_epochs=20,
    save_strategy = "no",
    lr_scheduler_type = "linear",
    metric_for_best_model= 'f1',
    adam_beta1 = 0.9,
    adam_beta2 = 0.999,
    adam_epsilon = 1e-08,
    disable_tqdm = False,
    do_eval = True,
    optim = "adamw_hf",
    seed = 42,
    weight_decay = 0.01,
    warmup_steps = 0,
    report_to = "none",
)
```

```python
import numpy as np
import evaluate

metric_acc = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric_acc.compute(predictions=predictions, references=labels)
```

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics = compute_metrics
)
```

```python
trainer.train()
```

```python
import random

raw_predictions = trainer.predict(test_dataset)
predicted_labels = np.argmax(raw_predictions.predictions, axis=1)

predicted_intents = [list(label_map.keys())[list(label_map.values()).index(label)] for label in predicted_labels]

sample_idx = random.randint(0, len(test_data) - 1)
sample_text = test_data[sample_idx]
sample_intent = predicted_intents[sample_idx]

print(f"Sample Text: {sample_text}")
print(f"Predicted Intent: {sample_intent}")
```

```python
from sklearn.metrics import classification_report

report = classification_report(test_labels, predicted_labels, target_names=list(label_map.keys()), output_dict=True)

accuracy = report['accuracy']
precision = report['macro avg']['precision']
recall = report['macro avg']['recall']
f1_score = report['macro avg']['f1-score']

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```
